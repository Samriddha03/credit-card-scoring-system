```python

# Get home folder
home = os.path.expanduser("~")
```

```python
# Source path of your model
src = os.path.join(home, "credit_risk_model.pkl")
```

```python
# Destination path (your project folder)
dst = r"D:\Desktop\credit-card-scoring-system\credit_risk_model.pkl"
```

```python
# Move the model
if os.path.exists(src):
    shutil.move(src, dst)
    print("Model moved successfully!")
else:
    print("Model file not found in home directory!")
```
